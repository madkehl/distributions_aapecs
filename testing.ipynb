```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_subplots(df, column):
    grouped = df[['participantID', column]].groupby('participantID')
    rowlength = round(grouped.ngroups/10)                         # fix up if odd number of groups
    fig, axs = plt.subplots(figsize=(20,40), 
                            nrows=10, ncols=rowlength,     # fix as above
                            gridspec_kw=dict(hspace=1)) # Much control of gridspec

    targets = zip(grouped.groups.keys(), axs.flatten())
    for i, (key, ax) in enumerate(targets):
        group = grouped.get_group(key)[column]
        ax.plot(group.reset_index(drop = True))
        ax.set_title('part_ID=%d'%key)
    ax.legend()
    plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)
    plt.show()
```
